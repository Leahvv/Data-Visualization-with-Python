```python
haiti = df_can.loc['Haiti',years]
haiti.head()
```

```python
haiti.plot()
```

```python
haiti.index = haiti.index.map(int)
haiti.plot(kind = 'line')

plt.title('Immigration from Haiti')
plt.xlabel('Year')
plt.ylabel('Number of Immigrats')

plt.show()
```

```python
haiti.plot(kind= 'line')

plt.title('Immigration from Haiti')
plt.xlabel('Years')
plt.ylabel('Number of Immigrants')

plt.text(2012, 6000,'2010 Earthquake')

plt.show()
```

```python
df_CI = df_can.loc[['China', 'India'],years]
df_CI.head()
```

```python
df_CI.plot(kind='line')
```

```python
df_CI = df_CI.transpose()
df_CI.head()
```

```python
df_CI.index = df_CI.index.map(int)
df_CI.plot(kind= 'line')

plt.title('Immigration from China and Idia')
plt.xlabel('Years')
plt.ylabel('Number of Immigrants')

plt.show()
```

```python
inplace= True
df_can.sort_values(by = 'Total',ascending = False, axis= 0, inplace = True)
top5 = df_can.head(5)

top5 = top5[years].transpose()

print(top5)
```

```python
top5.index = top5.index.map(int)
top5.plot(kind= 'line', figsize = (14,8))

plt.title('Immigration of Top 5')
plt.xlabel('Years')
plt.ylabel('Number of Immigrants')

plt.show()
```
